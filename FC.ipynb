```python
## FC
import numpy as np
from nilearn.connectome import ConnectivityMeasure, sym_matrix_to_vec

# --------------------------------------------------
# Load time-series data
# Each element in the list is an array with shape (T_i, n_ROIs)
# --------------------------------------------------

post_path = "path/to/time_series_list_24_postExtRest.npy"

time_series_list_post = list(
    np.load(post_path, allow_pickle=True)
)

# --------------------------------------------------
# Compute functional connectivity (Pearson correlation)
# --------------------------------------------------

correlation_measure = ConnectivityMeasure(kind="correlation")

# Result shape: (n_subjects, n_ROIs, n_ROIs)
correlation_matrices_post = correlation_measure.fit_transform(
    time_series_list_post
)

# --------------------------------------------------
# Convert symmetric matrices to feature vectors
# (upper triangular elements without the diagonal)
# --------------------------------------------------

fc_post = sym_matrix_to_vec(
    correlation_matrices_post,
    discard_diagonal=True
)

print("FC matrix shape:", fc_post.shape)
```
